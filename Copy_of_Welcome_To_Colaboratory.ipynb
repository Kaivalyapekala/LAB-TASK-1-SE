```
import matplotlib.pyplot as plt
import numpy as np
def calculate_temperature(time):
    a = 0.02
    b = 0.0
    c = 20.0
    temperature = a * (time ** 2) + b * time + c
    return temperature
time_values = np.arange(0, 24, 0.1)

temperature_values = [calculate_temperature(time) for time in time_values]
plt.plot(time_values, temperature_values, label='Temperature Model')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.title('Simple Weather Model Using Quadratic Equation (Parabola)')
plt.legend()
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

def calculate_temperature(time, a, b, c):
    return a * (time ** 2) + b * time + c

def plot_weather_model(time_values, temperature_values, label):
    plt.plot(time_values, temperature_values, label=label)

def get_coefficients_from_user():
    a = float(input("Enter the quadratic coefficient (a): "))
    b = float(input("Enter the linear coefficient (b): "))
    c = float(input("Enter the constant term (c): "))
    return a, b, c

def get_coefficients_from_file(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            a, b, c = map(float, lines)
            return a, b, c
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None

num_sets = int(input("Enter the number of sets of coefficients: "))

for i in range(num_sets):
    print(f"\nEnter coefficients for set {i + 1}:")


    input_method = input("Enter 'keyboard' for keyboard input or 'file' for reading from a file: ").lower()

    if input_method == 'keyboard':
        a, b, c = get_coefficients_from_user()
    elif input_method == 'file':
        filename = input("Enter the filename with coefficients: ")
        coefficients = get_coefficients_from_file(filename)

        if coefficients is not None:
            a, b, c = coefficients
        else:
            continue
    else:
        print("Invalid input method. Please enter 'keyboard' or 'file'.")
        continue
    time_values = np.arange(0, 24, 0.1)
    temperature_values = [calculate_temperature(time, a, b, c) for time in time_values]
    plot_weather_model(time_values, temperature_values, f'Temperature Model - Set {i + 1}')

# Show the plot
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.title('Weather Model Using Quadratic Equation')
plt.legend()
plt.grid(True)
plt.show()
```

```


import matplotlib.pyplot as plt
import numpy as np

def temperature_model(time, a, b, c):
    return a * time**2 + b * time + c

fixed = 0.5, 2, -1.0
user = [float(input(f"Enter the value for '{var}': ")) for var in ['a', 'b', 'c']]

time_values = np.linspace(1, 8, 100)

temp_fixed, temp_user = [temperature_model(time_values, *coeff) for coeff in [fixed, user]]

plt.plot(time_values, temp_fixed, label='Fixed Coefficients')
plt.plot(time_values, temp_user, label='User Input Coefficients')

for var, val in zip(['a', 'b', 'c'], user):
    plt.annotate(f'User Input: {var}={val}', xy=(0.5, 0.9 - 0.05 * user.index(val)), xycoords='axes fraction', fontsize=10)

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Weather Modeling with Quadratic Equation')
plt.legend()
plt.show()
```

## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

To execute the code in the above cell, select it with a click and then either press the play button to the left of the code, or use the keyboard shortcut "Command/Ctrl+Enter". To edit the code, just click the cell and start editing.

Variables that you define in one cell can later be used in other cells:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Colab notebooks allow you to combine **executable code** and **rich text** in a single document, along with **images**, **HTML**, **LaTeX** and more. ## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
